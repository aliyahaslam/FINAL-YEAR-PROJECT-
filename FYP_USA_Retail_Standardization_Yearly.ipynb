```python
import pandas as pd
```

```python
retail = pd.read_csv('/content/drive/MyDrive/USA Retail Clean Data/USA CLEAN RETAIL SALES YEARLY.csv')
```

```python
retail.columns
```

```python
retail = retail.drop(columns=["Unnamed: 7"])
retail = retail.rename(columns={"Unnamed: 0":"Category"})
```

```python
retail.head()
```

```python
retail_list = retail.melt(id_vars=["Category"], var_name = "Year", value_name="Sales Amount")
```

```python
retail_list["Year"] = pd.to_numeric(retail_list["Year"])
```

```python
retail_list.head()
```

```python
population = pd.read_csv('/content/drive/MyDrive/Random Documents/usa population 2.csv')
```

```python
population.head()
```

```python
retail_list = retail_list.merge(population, on="Year", how="left")
```

```python
retail_list.head()
```

```python
retail_list["Sales Amount"] = (
    retail_list["Sales Amount"].str.replace(",", "").astype(float)
)
```

```python
retail_list["Sales Amount per Capita"] = (retail_list["Sales Amount"] * 100)/retail_list["Population"]
```

```python
retail_list.head()
```

```python
retail_list.to_csv("retail_list.csv", index=False)
```
