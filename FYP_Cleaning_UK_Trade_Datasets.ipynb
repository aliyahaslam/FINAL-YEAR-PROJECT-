```python
import pandas as pd
```

```python
uk_yexports = pd.read_csv('/content/drive/MyDrive/UK Trade Clean Data/UK Annual Exports Clean.csv')
uk_yexports.head()
```

```python
uk_yexports.info()
```

```python
uk_yexports.describe()
```

```python
uk_yexports.isnull().sum()
```

```python
uk_yexports.columns
```

```python
population = pd.read_csv('/content/drive/MyDrive/Random Documents/ONS population.csv')
```

```python
population.head()
```

```python
uk_yexports = uk_yexports.melt(id_vars= ["Country Code", "Country Name"], var_name = "Year", value_name= "Annual Exports")
```

```python
uk_yexports["Year"] = pd.to_numeric(uk_yexports["Year"])
```

```python
uk_yexports.head()
```

```python
uk_yexports = uk_yexports.merge(population, on = "Year", how="left")
```

```python
uk_yexports.head()
```

```python
uk_yexports["Annual Exports per Capita"] = (uk_yexports["Annual Exports"] * 10000) / uk_yexports["Population "]
```

```python
uk_yexports.head()
```

```python
uk_yexports.to_csv('Standardized_uk_exports.csv', index = False)
```
