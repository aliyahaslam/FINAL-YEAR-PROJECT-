```python
fruit = pd.read_csv('/content/drive/MyDrive/USA Trade Clean Data/USA CLEAN IMPORT FRUIT, VEG, NUTS.csv')
```

```python
fruit.head()
```

```python
# prompt: Using dataframe fruit: Can I get a line chart of the the highest points and the lowest points for each year - make those two seperate lines as well

import altair as alt
import pandas as pd

# Group by year and find the max and min values
yearly_data = fruit.groupby('Year')['Total Imports'].agg(['max', 'min']).reset_index()

# Create a line chart for the maximum values
max_line = alt.Chart(yearly_data).mark_line(color='red').encode(
    x='Year',
    y='max'
)

# Create a line chart for the minimum values
min_line = alt.Chart(yearly_data).mark_line(color='blue').encode(
    x='Year',
    y='min'
)


# Combine the charts into one
chart = (max_line + min_line).resolve_scale(y='independent')


chart
```
