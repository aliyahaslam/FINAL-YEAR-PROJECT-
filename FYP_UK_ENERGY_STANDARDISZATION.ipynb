```python
electricity_list = electricity.melt(id_vars=["Primary or secondary electricity", "Generator type", "Technology"], var_name = "Year", value_name="Electricity Production (GWh()" )
```

```python
electricity_list["Year"] = pd.to_numeric(electricity_list["Year"])
```

```python
electricity_list.head()
```

```python
electricity_list = electricity_list.merge(population, on="Year", how="left")
```

```python
electricity_list.head()
```

```python
electricity_list.rename(columns={"Electricity Production (GWh()": "Electricity Production (GWh)"}, inplace=True)
```

```python
electricity_list["Electricity Production (GWh)"] = (
    electricity_list["Electricity Production (GWh)"].str.replace(",", "").astype(float)
)
```

```python
electricity_list["Electricity Production per Capita"] = (electricity_list["Electricity Production (GWh)"] * 1000) / electricity_list["Population "]
```

```python
electricity_list
```

```python
# prompt: Using dataframe electricity_list: Can I get a line chart with the highest points and lowest points of the electricity production per capita with two lines please

import altair as alt

# Calculate the maximum and minimum electricity production per capita for each year
electricity_max_min = electricity_list.groupby('Year')['Electricity Production per Capita'].agg(['max', 'min']).reset_index()

# Create a line chart with the maximum and minimum values
chart = alt.Chart(electricity_max_min).mark_line().encode(
    x='Year',
    y='max',
    color=alt.value('blue')
).properties(title="Maximum and Minimum Electricity Production per Capita")


chart2 = alt.Chart(electricity_max_min).mark_line().encode(
    x='Year',
    y='min',
    color=alt.value('red')
).properties(title="Maximum and Minimum Electricity Production per Capita")

chart + chart2

```

```python
electricitycb = electricitycb.merge(population, on="Year", how="left")
```

```python
electricitycb.columns = electricitycb.columns.str.strip()
```

```python
electricitycb["Electricity Commodity Balance"] = pd.to_numeric(electricitycb["Electricity Commodity Balance"], errors="coerce")
electricitycb["Population"] = pd.to_numeric(electricitycb["Population"], errors="coerce")
```

```python
electricitycb["Electricity Commodity Balance"] = pd.to_numeric(electricitycb["Electricity Commodity Balance"], errors="coerce")
```

```python
electricitycb["Electricity Commodity Balance"].fillna(0, inplace=True)
```

```python
electricitycb["Electricity Commodity Balance"].fillna(electricitycb["Electricity Commodity Balance"].mean(), inplace=True)
```

```python
electricitycb["Electricity Commodity Balance Per Capita"] = (electricitycb["Electricity Commodity Balance"] * 1000) / electricitycb["Population"]
```

```python
# prompt: Using dataframe electricitycb: can i get a line chart of the electricty commodity balance per capita where one line is the highest points

# Import necessary libraries
import altair as alt
import pandas as pd

# Create a new column for the highest points
electricitycb['Highest Point'] = electricitycb.groupby('Year')['Electricity Commodity Balance Per Capita'].transform('max')

# Create a line chart with two lines
chart = alt.Chart(electricitycb).mark_line().encode(
    x='Year',
    y='Electricity Commodity Balance Per Capita',
    color=alt.Color('Category', legend=alt.Legend(title='Category'))
)
chart2 = alt.Chart(electricitycb).mark_line(color='red').encode(
    x='Year',
    y='Highest Point',
)
final_chart = chart + chart2
final_chart

```
